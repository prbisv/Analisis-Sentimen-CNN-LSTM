```python
# membaca salah satu file JSON dengan pandas dataframe
df = pd.read_json('shopee_raw.json', lines=True)

# menampilkan data
df
```

```python
# membuat dataframe sementara untuk memilih fitur yang akan digunakan

data_shopee = pd.DataFrame()

# append tweet dan waktu dari dataframe lama ke dataframe baru

data_shopee['time'] = df['date']
data_shopee['tweet'] = df['rawContent']

data_shopee
```
